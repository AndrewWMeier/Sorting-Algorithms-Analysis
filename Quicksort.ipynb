<br>* random     => for generating random code
<br>
<br>* numpy      => for efficient array manipulation and some statistical operations, such as computing averages and standard deviations.

<br>* timeit     => It's used to measure the execution time of the Quick Sort algorithm for each array during the empirical analysis.
<br>

```python
import random
import numpy as np
import timeit
import matplotlib.pyplot as plt
```

<b>Step 2: Implement Quick Sort Algorithm</b>
<br>
<br> Define the Quick Sort algorithm

```python
def quicksort(arr):
    # Base case: return the array if its length is 1 or less
    if len(arr) <= 1:
        return arr
    else:
        # Choose the pivot as the first element
        pivot = arr[0]
        # Divide the array into two subarrays: elements <= pivot and elements > pivot
        left = [x for x in arr[1:] if x <= pivot]
        right = [x for x in arr[1:] if x > pivot]
        # Recursively apply quicksort on the left and right subarrays
        return quicksort(left) + [pivot] + quicksort(right)
```

<b>Step 3: Generate Random Arrays</b>
<br>
<br>Create a function to generate random arrays


```python
def generate_arrays(arr_size, num_arrays, integer_range):
    arrays = []
    for i in range(num_arrays):
        # Generate an array with random integers in the specified range
        arr = [random.randint(0, integer_range) for _ in range(arr_size)]
        arrays.append(arr)
    return arrays
```

<b>Step 4: Empirical Analysis Function<b>
<br>
<br>Write a function to perform empirical analysis


```python

def empirical_analysis(array_sizes, num_arrays, integer_range):
    execution_times = []
    for size in array_sizes:
        # Generate random arrays for the given size
        test_data = generate_arrays(size, num_arrays, integer_range)
        start_time = timeit.default_timer()
        # Apply quicksort to each array and measure execution time
        for arr in test_data:
            quicksort(arr)
        execution_times.append(timeit.default_timer() - start_time)
    return execution_times
```

<b>Step 5: User input to Perform Analysis </b>
<br>
<br>User inputs parameters for analysis and empirical analysis function is called after


```python
# Get user input for array sizes
array_sizes_input = input("Enter array sizes (comma-separated): ")
array_sizes = [int(size) for size in array_sizes_input.split(",")]

# Get user input for the number of arrays
num_arrays = int(input("Enter the number of arrays: "))

# Get user input for the integer range
integer_range = int(input("Enter the integer range: "))

# Perform the empirical analysis
execution_times = empirical_analysis(array_sizes, num_arrays, integer_range)

# Visualize Results
plt.plot(array_sizes, execution_times, marker='o')
plt.title("Quick Sort Empirical Analysis")
plt.xlabel("Array Size")
plt.ylabel("Execution Time (s)")
plt.show()
```

```python
# Calculate additional statistics using NumPy
mean_execution_time = np.mean(execution_times)
std_dev_execution_time = np.std(execution_times)

print(f"Mean Execution Time: {mean_execution_time} seconds")
print(f"Standard Deviation of Execution Time: {std_dev_execution_time} seconds")
```
